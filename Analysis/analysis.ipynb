```python
fig = plt.figure()
ax = fig.add_subplot(111,projection = '3d')

x=df["Height of tree"]
y=df["No of unique propositional atoms"]
z=df["Time(ms)"]

ax.set_xlabel("Height of tree")
ax.set_ylabel("No of propositional atoms")
ax.set_zlabel("Time(ms)")

ax.scatter(x,y,z)
plt.show()
```
